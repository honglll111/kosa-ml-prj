## 제주 방언

```python
# 제주 방언
import os
from tqdm import tqdm

path = 'C:/Users/KOSA/Desktop/Korean/한국어 방언 발화(제주도)/Training/[라벨]제주도_학습용데이터_1/'
file_list = os.listdir(path)
files = [file for file in file_list if file.endswith('.json')] # 파일명 끝이 .json인 경우

import json
dialect_list = []
for file in tqdm(files):
    with open((path+file), encoding='utf-8-sig') as f:
        data = json.load(f)
        utterance = data['utterance']
        for sentence in utterance:
            if sentence['standard_form'] == sentence['dialect_form']:
                pass
            else:
                dialect_list.append(sentence['dialect_form'])
```

```python
import numpy as np
dialect_array = np.array(dialect_list)
dialect_array = dialect_array.reshape(-1, 1); dialect_array

import pandas as pd
df = pd.DataFrame(dialect_array)
df.columns = ['sentence']
df['file'] = 'jeju'

df.to_csv('dialect_jeju.csv', index=False, encoding='utf-8-sig')
```

### 정규표현식

```python
import pandas as pd

df = pd.read_csv('dialect_all.csv')
```

```python
# 결측치 있어서 제거함
df = df.dropna()
```

```python
# 1. 끊어진 단어(단어가 불완전하게 발화된 경우) - 18p
# ex) -학- 학교 아니 유치원에
# text1 = r'-[가-힣\s]*\-'  # 한글, 공백이 포함되었을 경우

# 2. 담화 표지 = 22p
# ex) 응~ 얼마만인지 모르겠네 
# test2 = r'[가-힣]~'

### 3. 잘 들리지 않는 부분(이거 안함) -> (())만 제거
### ex) 그거 진짜 ((xx해야)) 되겠더라
### test3 = r'\(\(.*?\)\)'

# 4. 말소리를 제외한 기타 소리들 - 23p
# ex) 그러니께 그게 문제여 {laughing}
# text4 = r'\{[a-zA-Z\s]*\}'

# 5. 말소리를 제외한 기타 소리들(전사), 개인 정보의 보호(전사)
# ex) 하나하나에 몽땅 이름들을 써줍디다 아버지가 @인물5 @인물6 @인물7 써주면 다
# text5 = r'@[가-힣0-9]+'

# 6) 개인 정보의 보호-1
# ex) 근데 이제 &company-name1& 코로나 때문에 영화 산업들이 쪼끔 약간 쫌
# text6 = r'&[a-z0-9-]+&'

# 7. 문장 부호, 영어, 숫자 제외
# ex) 그냥 그냥 먹어도 문작문작 아무치도 안해네 그냥.
# text7 = r'[^가-힣* ]+'
```

```python
import re
from tqdm import tqdm

def preprocessing(text):
    text = re.sub(r'-[가-힣\s]*\-', ' ', text)
    text = re.sub(r'[가-힣]~', ' ', text)
    text = re.sub(r'\(\(.*?\)\)', ' ', text)
    text = re.sub(r'\{[a-zA-Z\s]*\}', ' ', text)
    text = re.sub(r'@[가-힣0-9]+', ' ', text)
    text = re.sub(r'&[a-z0-9-]+&', ' ', text)
    text = re.sub(r'[^가-힣a-z0-9* ]+', ' ', text)

    return text
```

```python
# map을 통해 전처리 일괄 적용
df["sentence"] = df["sentence"].map(preprocessing)
```

```python
# 결과 CSV 파일로 저장
df.to_csv('dialect_clear.csv', index=False, encoding='utf-8-sig')
```

### 지역별 문장 30000개 랜덤으로 선택

```python
def ramdom_dialect_ramdom_30000():
    df[df['file'] == 'chungcheong'].sample(n=30000, random_state=42)
    df[df['file'] == 'jeju'].sample(n=30000, random_state=42)
    df[df['file'] == 'jeolla'].sample(n=30000, random_state=42)
    df[df['file'] == 'kangwon'].sample(n=30000, random_state=42)
    df[df['file'] == 'kyungsang'].sample(n=30000, random_state=42)
```

```python
def dialect_ramdom_30000():
    # 각 방언에 대해 30000개씩 랜덤 샘플 선택
    random_samples = []
    dialects = ['chungcheong', 'jeju', 'jeolla', 'kangwon', 'kyungsang']
    
    for dialect in dialects:
        sample = df[df['file'] == dialect].sample(n=30000, random_state=42)
        random_samples.append(sample)
    
    # 모든 샘플을 하나의 데이터프레임으로 합치기
    combined_df = pd.concat(random_samples)
    
    # CSV 파일로 저장 (encoding='utf-8-sig' 옵션 추가)
    combined_df.to_csv('dialect_random_30000.csv', index=False, encoding='utf-8-sig')

# 함수 호출하여 실행
dialect_ramdom_30000()
```

###  형태소 분석을 통한 조사, 구두점 제거

```python
from pecab import PeCab
pecab = PeCab()
```

```python
# 조사, 구두점 제거 조사, 구두점 제거
def remove_josa_puct(corpus_txt):
    pecab_pos = pecab.pos(corpus_txt, drop_space=False)
    pos_after = []
    for txt, pos in pecab_pos:
        if not pos.startswith("J") and pos not in ("SF"):
            pos_after.append(txt)
    return "".join(pos_after)
```

```python
from tqdm import tqdm
tqdm.pandas()  # 진행사항 보여줌

df1["sentence"] = df1["sentence"].progress_map(remove_josa_puct)
df1.to_csv("dialect_pos.csv", index=False, encoding='utf-8-sig')
```
